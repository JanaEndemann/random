```python
import pandas as pd
```

# Data

```python
df = pd.read_excel('afschrift.xlsx')
```

```python
df.loc[df['Debet / Credit'] == 'Debet','Bedrag'] *= -1
```

```python
df.head()
```

```python
interesting_cols = ['Boekdatum','Bedrag','Naam tegenrekening','Omschrijving']
```

# Inkomende

## Weekend

```python
betaald_voor_weekend = df.loc[(df.Bedrag % 90 == 0) & (df.Bedrag > 0)][interesting_cols]
betaald_voor_weekend['aantal_personen'] = betaald_voor_weekend.Bedrag /90
betaald_voor_weekend['voor'] = 'weekend'
betaald_voor_weekend
```

```python
betaald_voor_weekend.aantal_personen.sum()
```

## Dagje

```python
betaald_voor_dag = df[(df.Bedrag % 35 == 0) & (df.Bedrag > 0)][interesting_cols]
betaald_voor_dag['aantal_personen'] = betaald_voor_dag.Bedrag /35
betaald_voor_dag['voor'] = 'dagje'
betaald_voor_dag
```

```python
betaald_voor_dag.aantal_personen.sum()
```

## Bas

```python
betaald_bas = df[df['Naam tegenrekening'] == 'Hr B Joustra'][interesting_cols]
betaald_bas['aantal_personen'] = 1.
betaald_bas['voor'] = 'extra'
betaald_bas
```

```python
betaald_allen = pd.concat([betaald_voor_weekend,betaald_voor_dag, betaald_bas], axis = 0, sort = False)
betaald_allen.to_csv('betaald.csv', index = False)

betaald_neat = betaald_allen.copy()
for col in betaald_neat.columns:
    pad_length = max(betaald_neat[col].apply(str).apply(len).max(), len(col))
    betaald_neat[col] = [str(val).ljust(pad_length) for val in list(betaald_neat[col])]

betaald_neat[['Naam tegenrekening','aantal_personen','voor']].to_csv('betaald_kort.csv', index = False, sep = ':')
betaald_neat
```

```python
betaald_allen.Bedrag.sum()
```

```python
df_totaal = pd.DataFrame(columns = ['betaald_door','betaald_aan','bedrag'])
```

```python
def add_to_total(omschrijving, betaald_door, betaald_aan, bedrag):
    df_totaal.loc[omschrijving,'betaald_door'] = betaald_door
    df_totaal.loc[omschrijving,'betaald_aan'] = betaald_aan
    df_totaal.loc[omschrijving,'bedrag'] = bedrag
    return df_totaal
```

```python
df_totaal = add_to_total('overnachtingen','lui','Jana',betaald_allen.Bedrag.sum())
df_totaal
```
