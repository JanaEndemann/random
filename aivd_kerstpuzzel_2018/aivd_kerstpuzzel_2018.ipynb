```python
import string
```

```python
def get_letter(n):
    abc = string.ascii_lowercase
    return abc[n-1]
```

```python
def show_word(a,b,c):
    s = get_letter(a+3*b+c+12)
    s += get_letter(a+2*b+4)
    s += get_letter(b+c+8)
    s += get_letter(a+b)
    s += get_letter(b+4)
    s += get_letter(c+4)
    s += get_letter(a)
    s += get_letter(b)
    s += get_letter(4)
    s += get_letter(c)
    return s
```

```python
a_range = range(1,11)
b_range = range(1,5)
c_range = range(1,11)
```

```python
l = ['a,b,c,word']
for a in a_range:
    for b in b_range:
        for c in c_range:
            if (a+3*b+c+12-1) < 26:
                l.append('{}, {}, {}, {}'.format(get_letter(a),get_letter(b),get_letter(c),show_word(a,b,c)))
```

```python
l[:5]
```

```python
def print_to_csv(write_file,l):
    with open(write_file, "w") as output:
        for line in l:
            output.write(line + '\n')
```

```python
print_to_csv("aivd_4.csv",l)
```
